# histogram and displot in seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.displot(tips["size"])
```

```python
sns.displot(tips["tip"])
```

```python
sns.displot(tips["total_bill"])
```

```python
sns.displot(tips["total_bill"] , bins = 10)
```

```python
sns.displot(tips["total_bill"] , bins = 100)
```

```python
sns.displot(tips["total_bill"]  ,kde = True)
```

```python
sns.displot(tips["total_bill"]  ,kde =False  )
```

```python
sns.displot(tips["total_bill"]  ,rug = True)
```

```python
sns.displot(tips["total_bill"]  ,color ='r'  )
```

```python
sns.displot(tips["total_bill"]  ,color ='r',  label = "histogram")
plt.title("Histogram of tatal bill" , fontsize = 20)
plt.legend()
```

```python
plt.figure(figsize = (15,9))

sns.set( style = "dark")

sns.displot(tips["total_bill"]  ,color ='r',  label = "histogram")
plt.title("Histogram of tatal bill" , fontsize = 20)
plt.legend()
plt.show()
```

```python
tips.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.displot(tips["total_bill"], bins = bins )

plt.xticks(bins)

plt.title("Histogram of tatal bill" , fontsize = 20)
plt.show()
```
