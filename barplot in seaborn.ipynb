
# barplot in seaborn in python & bar Graph

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.barplot(x = tips.total_bill)
```

```python
sns.barplot(y = tips.total_bill ,x = tips.day)
```

```python
sns.barplot(y = tips.total_bill ,x = tips.day, hue =tips.sex)
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex' ,data = tips)
```

```python
order = [ 'Sun' , 'Fri' , 'Thur' ,'Sat']

sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex' ,data = tips ,order = order)
```

```python
hue_order = [ 'Female' ,'Male']

sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex' , hue_order = hue_order ,data = tips )
```

```python


sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , estimator=np.mean)
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , estimator=np.max)
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , estimator=np.max ,ci = 100)
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , estimator=np.max ,ci = 100 ,color = 'g')
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , estimator=np.max ,ci = 100 ,palette = 'hot')
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips ,saturation= 8)
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , errcolor= '1.0')
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , errcolor= '1.0', errwidth = 10)
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , errcolor= '1.0', errwidth = 10 ,capsize= 0.1
           )
```

```python
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , dodge= False)
```

```python
kwargs = {'alpha' : 0.9 , 'linestyle' : ':' , 'linewidth': 5 , 'edgecolor' : 'k'}
sns.barplot(y = 'total_bill' ,x = 'day', hue = 'sex'  ,data = tips , dodge= False ,**kwargs)
```

```python
sns.set()

kwargs = {'alpha' : 0.9 , 'linestyle' : ':' , 'linewidth': 5 , 'edgecolor' : 'k'}
sns.barplot(y = 'total_bill' ,x = 'day',   data = tips , dodge= False ,**kwargs)
```

```python
ax =sns.barplot(y = 'total_bill' ,x = 'day',   data = tips , dodge= False ,**kwargs)

ax.set(title = "data frame",
      xlabel = "days ",
      ylabel = "total bills" )
```

```python
plt.figure(figsize = (15,9))
ax =sns.barplot(y = 'total_bill' ,x = 'day',   data = tips , dodge= False ,**kwargs)

ax.set(title = "data frame",
      xlabel = "days ",
      ylabel = "total bills" )
```
