# Scallter plot in seaborn in python

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tit = sns.load_dataset('titanic')
```

```python
tit
```

```python
sns.scatterplot( x = "age" , y = 'fare', data = tit)
```

```python
sns.scatterplot( x = "age" , y = 'fare', data = tit, hue = 'sex')
```

```python
sns.scatterplot( x = "age" , y = 'fare', data = tit, hue = 'sex' ,style ='who')
```

```python
sns.scatterplot( x = "age" , y = 'fare', data = tit, hue = 'sex' ,style ='who' ,size = "who")
```

```python
sns.scatterplot( x = "age" , y = 'fare', data = tit, hue = 'sex' ,style ='who' ,size = "who" ,sizes = (100 ,400))
```

```python
plt.figure(figsize= (16,9))
sns.scatterplot( x = "age" , y = 'fare', data = tit, hue = 'sex' ,style ='who' ,size = "who" ,sizes = (100 ,400))
```

```python
plt.figure(figsize= (16,9))
sns.scatterplot( x = "who" , y = 'fare', data = tit, hue = 'alive' ,style ='alive' ,size = "who" ,sizes = (100 ,400))
```

```python
plt.figure(figsize= (16,9))
sns.scatterplot( x = "who" , y = 'fare', data = tit, hue = 'alive' ,style ='alive' ,size = "who" ,sizes = (100 ,400) ,
               palette = "inferno_r" , alpha =.7)
```
